DATA ANALYSIS

```python
#identifying outliers in price

df=data[data['price']<1500].copy()
sns.boxplot(data=df,x='price',color='g')
```

```python
#price distribution

plt.figure(figsize=(10,6))
sns.histplot(df['price'],bins=80,color='green')
plt.title('Distribution of price')
plt.ylabel('Freaquency')
plt.show()
```

```python
#Distribution of Availability_365


plt.figure(figsize=(10,6))
sns.histplot(df['availability_365'],bins=30,color='green')
plt.title('Availability 365_Distribution')
plt.ylabel('Freaquency')
plt.show()
```

```python
#total number of each neighbourhood group


ngcount = data.groupby('neighbourhood_group')['neighbourhood_group'].count()
print(ngcount)
```

```python
#rename the column neighbourhoodgroup into borough

df.rename(columns={'neighbourhood_group':'borough'},inplace = True)
print(df.head(1))
```

```python
#Average price according to different borough


df.groupby(by='borough')['price'].mean()
```

```python
#price per bed

df['price_per_bed'] = df['price'] / df['beds']
df.head()
```

```python
#Average price per bed

avg_price = df.groupby('borough')['price_per_bed'].mean()
print(avg_price)
```

```python
#Price dependancy on neighbourhood groups(borough)

plt.figure(figsize=(10,6))
sns.barplot(data=df,x='borough',y='price',hue='room_type')
```

```python
# How many listings are there for each 'room_type'?

data['room_type'].value_counts()
```

```python
#maximum price by room type 

maxprice =data.groupby('room_type')['price'].max()
print(maxprice)
plt.bar(maxprice.index,maxprice.values,color='g')
plt.xlabel('Room type',color= 'r')
plt.ylabel('Max price',color='r')

```

```python
#Average price according to room type

avgprice = data.groupby('room_type')['price'].mean()
print(avgprice)
plt.bar(avgprice.index,avgprice.values,color='g')
plt.xlabel('Room type',color= 'r')
plt.ylabel('Avg price',color='r')
plt.ylim(100,500)
```

```python
# How does price correlate with the number of reviews?


sns.scatterplot(y='price', x='number_of_reviews',hue='borough', data=df)
plt.title('Locality and review dependancy')
plt.ylabel('Price')
plt.xlabel('Number of Reviews')
plt.show()
```

```python
## Number of hosts per neighbourhoodgrp


hosts_per_borough = df.groupby('borough')['host_id'].nunique()
print(hosts_per_borough) 
```

```python
# Count the number of rows with "Non licensed"


non_licensed_count = data[data['license'] == 'No License'].shape[0]
print(f"Number of Non Licensed listings: {non_licensed_count}")

```

```python
#top 10 avg rating neighbourhoods 

data['rating'] = pd.to_numeric(data['rating'], errors='coerce')
filtered_data = data.dropna(subset=['rating'])
avg_rating_per_neighborhood = (
    filtered_data.groupby('neighbourhood')['rating']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

print(avg_rating_per_neighborhood)
```

```python
#What are the top 10 hosts with the most listings?


top_hosts = data['host_name'].value_counts().head(10)
top_hosts.plot(kind='bar', color='skyblue')
plt.title('Top 10 Hosts with the Most Listings')
plt.xlabel('Host Name',color='r')
plt.ylabel('Number of Listings',color='r')
plt.xticks(rotation=45)
plt.show()
```

```python
#Geographical distribution of airbnb listing


plt.Figure(figsize=(10,7))
sns.scatterplot(data=df,x='longitude',y='latitude',hue='room_type')
```

```python
#pairwise relationships between multiple numeric variables.

sns.pairplot(data=df,vars=['price','min_nights','number_of_reviews','availability_365'],hue='room_type')
```

```python
#Heatmap(visualizing relationships between variables in a matrix format)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')  

corelation = df[['longitude','latitude','price','min_nights','number_of_reviews','reviews_per_month','availability_365','beds','rating']].corr()
corelation
sns.heatmap(data=corelation,annot=True,cmap='coolwarm')
```

Key Findings and Insights

1) Price Distribution Insights:

   * The majority of Airbnb listings are priced within an affordable range, but there are some high-price outliers,       especially in Manhattan.
   
   * Entire homes/apartments cost significantly more than private or shared rooms.

2) Neighbourhood Group Trends:

   * Brooklyn and Manhattan have the highest number of listings, making them key areas for Airbnb activity.

3) Host Activity & Market Trends:

   * A significant number of hosts have multiple listings, indicating professional hosting in some areas.

   * Understanding licensed vs. non-licensed listings helps identify legal and compliance issues.

4) Guest Experience & Ratings:

   * Higher-rated properties tend to have reasonable pricing and better amenities.   
   * The analysis reveals an inverse relationship between price and the number of reviews, indicating that lower-priced properties tend to receive more reviews. This suggests that budget-friendly listings attract a higher number of guests, leading to more frequent feedback and engagement.


5) Enhancing Availability & Occupancy

   * Some boroughs have low availability (availability_365 < 100 days/year), limiting booking opportunities.
   * Increasing availability and optimizing minimum night stays can attract more guests.

5) Latitude & Longitude Correlation:

   * Most of the entire home/apt are situated in (-40.75,-74.05)

6) Future Data-Driven Decisions:

   * Implementing machine learning models could improve price prediction and occupancy rates.
